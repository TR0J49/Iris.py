```python
numeric_data = data.select_dtypes(include='number')


correlation_train = numeric_data.corr()
print("Correlation Matrix for Training Data:")
print(correlation_train)


numeric_data = data.select_dtypes(include='number')


correlation_test = numeric_data.corr()
print("\nCorrelation Matrix for Test Data:")
print(correlation_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation_train, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Training Data')
plt.show()
```

```python
numerical_columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 4, i)
    sns.histplot(data[col], kde=True, color='blue', bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_classifier.fit(x_train, y_train)
```

```python
predictions = knn_classifier.predict(x_test)
```

```python
print(predictions)
```

```python
print("Training accuracy Score is:", accuracy_score(y_train, knn_classifier.predict(x_train)))
```

```python
print("Testing accuracy Score is:", accuracy_score(y_test, predictions))
```

```python
print("Training Confusion Matrix is:\n", confusion_matrix(y_train, knn_classifier.predict(x_train)))
```

```python
print("Testing Confusion Matrix is:\n", confusion_matrix(y_test, predictions))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), predictions, color='green', alpha=0.5, label='Predicted Class')
plt.title('Predicted Classes (KNN Classifier)')
plt.xlabel('Sample Index')
plt.ylabel('Predicted Class')
plt.legend()
plt.show()
```
